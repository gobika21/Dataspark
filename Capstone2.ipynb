# Product csv to sql

```python
import pandas as pd
products = pd.read_csv(r"./data/Products.csv")
products
```

```python
products.isna().sum()
```

```python
products['Unit Cost USD'] = products['Unit Cost USD'].replace(r'[\$,]', '', regex=True).astype(float)
products['Unit Price USD'] = products['Unit Price USD'].replace(r'[\$,]', '', regex=True).astype(float)
products
```

```python
col = list(products.columns)
new_col = [c.strip().replace(' ', '_').replace('-', '').lower() for c in col]
products.columns = new_col
products
```

```python
products.to_csv('./data/products_frmt.csv', index=False)
```

```python
import pandas as pd
import mysql.connector

df = pd.read_csv(r'./data/products_frmt.csv')

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678",
    database='A2_DataSpark'
)
cursor = conn.cursor()
query = '''CREATE TABLE if not exists ProductTable(
    productkey BIGINT PRIMARY KEY,
    product_name VARCHAR(255),
    brand VARCHAR(255),
    color VARCHAR(100),
    unit_cost_usd FLOAT,
    unit_price_usd FLOAT,
    subcategorykey BIGINT,
    subcategory VARCHAR(255),
    categorykey BIGINT,
    category VARCHAR(255)
    )'''
cursor.execute(query)

insert_query = """
INSERT INTO ProductTable (productkey, product_name, brand, color, unit_cost_usd, unit_price_usd,
       subcategorykey, subcategory, categorykey, category
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Insert data into MySQL
for i, row in df.iterrows():
    cursor.execute(insert_query, tuple(row))
conn.commit()

# Close the connection
cursor.close()
conn.close()
```

# Sales csv to sql

```python
import pandas as pd
sales = pd.read_csv(r"./data/Sales.csv")
sales
```

```python
sales.isna().sum()
```

```python
sales.dropna(inplace=True)
```

```python
sales
```

```python
col = list(sales.columns)
new_col = [c.strip().replace(' ', '_').replace('-', '').lower() for c in col]
sales.columns = new_col
sales
```

```python
sales['order_date'] = pd.to_datetime(sales['order_date'], format='%m/%d/%Y', errors='coerce')
sales['delivery_date'] = pd.to_datetime(sales['delivery_date'], format='%m/%d/%Y', errors='coerce')
sales
```

```python
import pandas as pd
import mysql.connector

df = pd.read_csv(r'./data/sales_frmt.csv')

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678",
    database='A2_DataSpark'
)
cursor = conn.cursor()
query = '''CREATE TABLE if not exists SalesTable (
    order_number INT,
    line_item INT,
    order_date DATETIME,
    delivery_date DATETIME,
    customerkey INT,
    storekey INT,
    productkey INT,
    quantity INT,
    currency_code VARCHAR(10),
    PRIMARY KEY (order_number, line_item)
    )'''
cursor.execute(query)

insert_query = """
INSERT INTO SalesTable (order_number, line_item, order_date, delivery_date, customerkey, storekey,
       productkey, quantity, currency_code
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Insert data into MySQL
for i, row in df.iterrows():
    cursor.execute(insert_query, tuple(row))
conn.commit()

# Close the connection
cursor.close()
conn.close()
```

# Stores csv to sql

```python
stores.isna().sum()
```

```python
stores.dropna(inplace=True)
```

```python
col = list(stores.columns)
new_col = [c.strip().replace(' ', '_').replace('-', '').lower() for c in col]
stores.columns = new_col

stores['open_date'] = pd.to_datetime(stores['open_date'], format='%m/%d/%Y', errors='coerce')
stores['order_date'] = pd.to_datetime(stores['order_date'], format='%m/%d/%Y', errors='coerce')
stores
```

```python
import pandas as pd
import mysql.connector

df = pd.read_csv(r'./data/store_frmt.csv')

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678",
    database='A2_DataSpark'
)
cursor = conn.cursor()
query = '''CREATE TABLE if not exists StoreTable (
    storekey INT,
    country VARCHAR(255),
    state VARCHAR(255),
    square_meters FLOAT,
    open_date DATETIME,
    order_date DATETIME,
    PRIMARY KEY (storekey)
    )'''
cursor.execute(query)

insert_query = """
INSERT INTO StoreTable (storekey, country, state, square_meters, open_date, order_date
    ) VALUES (%s, %s, %s, %s, %s, %s)
"""

# Insert data into MySQL
for i, row in df.iterrows():
    cursor.execute(insert_query, tuple(row))
conn.commit()

# Close the connection
cursor.close()
conn.close()
```
